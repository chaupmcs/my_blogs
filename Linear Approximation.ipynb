Sometimes we don't need the exact value of $f(x)$. In this case, we can approximate the value of $g(x)$ based on a simple linear function $f(x)$ if the error is low enough. 

Let's say we have a tangent line [1] of $g(x)$ at the point $x = x1$. Give a near point called $x2$ as in the picture above.

We want calculate approximately $g(x2)$ by getting the value of $f(x2)$. Intuitively, we see that:

```python
plt.figure(figsize=(10, 15))

img = mpimg.imread('./image/linear_estimation_2.jpg')
imgplot = plt.imshow(img)
plt.show()
```

So, to calculate $g(x2)$, we can calculate $f(x1) + f'(x1)*(x2-x1)$.

Example, we want to estimate $ \sqrt{16.3} $ and we already know that  $ \sqrt{16} = 4 $.
From the formula, we have $\sqrt{16.3} = \sqrt{16} + 0.3/8 = 4.0375$

It's so close to the true value of $\sqrt{16.3} = 4.03732584763727$
